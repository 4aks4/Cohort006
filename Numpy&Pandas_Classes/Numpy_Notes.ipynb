```python
import pandas as pd

# create a DataFrame with some sample data
data = {'group': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
        'value': [10, 20, 30, 40, 50, 60, 70, 80],
        'value2':[90, 100, 110, 120, 130, 140, 150, 160]}
df = pd.DataFrame(data)

# use groupby to group data by 'group'
grouped = df.groupby('group')

# use filtering to remove rows where 'value' is less than 30
filtered = grouped.filter(lambda x: x['value'].mean() >= 30)

# use transformation to calculate the Z-score of 'value' for each group
transformed = filtered.groupby('group')['value'].transform(lambda x: (x - x.mean()) / x.std())

# use aggregation to calculate the mean value for each group
aggregated = transformed.groupby(filtered['group']).mean()

# print the result
print(aggregated)
```
